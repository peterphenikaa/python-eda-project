> **EDA** (*Exploratory Data Analysis* — *Phân tích khám phá dữ liệu*) là bước **nhìn, thống kê và hiểu dữ liệu trước khi huấn luyện model**: xem cấu trúc bảng, phân phối nhãn, giá trị thiếu, độ dài văn bản, mất cân bằng class...

```python
import pandas as pd

DATA_PATH = "data/tiki_reviews.csv"
df = pd.read_csv(DATA_PATH, encoding="utf-8-sig")

print(f"Số mẫu: {len(df):,}")
print(f"Số cột: {df.shape[1]}")
print(f"Các cột: {list(df.columns)}")
```

```python
print("=== Phân phối nhãn (label) ===")
label_counts = df["label"].value_counts()
display(label_counts.to_frame("count"))

print("\n=== Tỉ lệ % ===")
display((label_counts / len(df) * 100).round(2).to_frame("percent"))
```

### Nhận xét sơ bộ (Ngày 1)

- **Nguồn:** Đánh giá sản phẩm Tiki (tiếng Việt), repo [quangvh23/tiki-dataset-sentiment-classification](https://github.com/quangvh23/tiki-dataset-sentiment-classification).
- **Quy mô:** 13.652 mẫu, **3 nhãn** (`tieu_cuc`, `trung_lap`, `tich_cuc`) — đáp ứng yêu cầu đề (≥5.000 mẫu, ≥3 nhãn).
- **Ánh xạ nhãn:** rating 20/40 → tiêu cực, 60 → trung lập, 80/100 → tích cực.
- **Mất cân bằng (imbalanced):** lớp `tich_cuc` chiếm đa số (~85%), `tieu_cuc` rất ít (~2.6%). Model dễ **thiên về dự đoán lớp đa số** → accuracy cao nhưng recall lớp thiểu số kém. Cần dùng `stratify` khi split và cân nhắc `class_weight`/SMOTE ở Phần E.
