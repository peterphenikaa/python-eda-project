## A3 — Tiền xử lý văn bản (preprocess_text)

**Mục tiêu:** Xây dựng hàm làm sạch dữ liệu văn bản trước khi đưa vào mô hình học máy.

````python
import re
import string
import unicodedata
import nltk
from nltk.tokenize import word_tokenize

# Nếu máy chưa tải gói punkt của NLTK, hãy bỏ comment dòng dưới để tải về:
nltk.download('punkt')
nltk.download('punkt_tab')

# Danh sách stopwords tiếng Việt cơ bản (Do NLTK chưa hỗ trợ sẵn tiếng Việt)
vietnamese_stopwords = set(['và', 'là', 'của', 'có', 'trong', 'với', 'cho', 'một', 'những', 'để', 'như', 'nhưng', 'khi', 'các', 'cái', 'này', 'thì', 'rằng', 'bằng', 'từ', 'rất', 'quá', 'cũng', 'đã', 'được', 'lại', 'thêm', 'nhé', 'nha', 'ạ', 'ơi', 'đi'])

def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    # 1. Lowercase (Viết thường)
    # Chuẩn hóa Unicode (NFC) để sửa lỗi chữ tiếng Việt bị tách rời
    text = unicodedata.normalize('NFC', text)
    # 1. Lowercase (Viết thường)
    text = text.lower()

    # 2. Xóa HTML tags
    text = re.sub(r'<[^>]+>', ' ', text)

    # 3. Xóa URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 4. Xóa ký tự đặc biệt và dấu câu
    text = re.sub(r'[^\w\s]', ' ', text)

    # 5. Tokenize (Tách từ - Ở đây dùng word_tokenize của NLTK)
    tokens = word_tokenize(text)

    # 6. Xóa Stopwords
    tokens = [word for word in tokens if word not in vietnamese_stopwords]

    # 7. Stemming/Lemmatization
    # Tiếng Việt là ngôn ngữ đơn lập, không biến hình như tiếng Anh (không thêm s/ed/ing...)
    # Nên Stemming và Lemmatization gần như KHÔNG áp dụng cho tiếng Việt.
    # Để minh họa theo quy trình chuẩn, ta chỉ tiến hành nối các từ lại sau khi làm sạch.

    return " ".join(tokens)

# Áp dụng hàm preprocess_text để tạo cột text_clean
df['text_clean'] = df['text'].apply(preprocess_text)

# Xem kết quả trước và sau khi tiền xử lý
display(df[['text', 'text_clean']].head())
````

### Trả lời câu hỏi A3(a) và A3(b)

**A3(a). Tại sao phải xóa stopwords? Cho 5 ví dụ.**
- **Tại sao:** Stopwords (từ dừng) là những từ xuất hiện với tần suất rất cao trong ngôn ngữ nhưng lại mang rất ít hoặc không mang ý nghĩa để phân loại cảm xúc hay chủ đề (ví dụ: các giới từ, liên từ, đại từ). Việc xóa chúng giúp:
  1. Giảm nhiễu (noise) cho mô hình, giúp mô hình tập trung vào những từ khóa quan trọng mang tính cảm xúc (VD: "tuyệt_vời", "tệ", "hỏng").
  2. Giảm đáng kể số lượng chiều của không gian đặc trưng (số lượng features), giúp mô hình huấn luyện nhanh hơn và tránh overfitting.
- **5 ví dụ stopwords tiếng Việt:** "và", "là", "của", "thì", "mà".

**A3(b). So sánh Stemming và Lemmatization.**
- **Stemming (Cắt rễ từ):** Là thuật toán cắt bỏ phần hậu tố/tiền tố (như -ing, -ed, -s trong tiếng Anh) để đưa từ về dạng gốc rễ (stem) một cách thô bạo bằng các luật (rules) xử lý chuỗi. Dạng gốc này có thể không phải là một từ có nghĩa trong từ điển (VD: "caring" → "car", "studies" → "studi"). Tốc độ xử lý rất nhanh nhưng độ chính xác kém.
- **Lemmatization (Chuẩn hóa từ):** Là quá trình đưa từ về dạng cơ sở (lemma) dựa trên ngữ pháp và từ điển thực tế. Quá trình này hiểu ngữ cảnh của từ (VD: "better" → "good", "running" → "run"). Tốc độ xử lý chậm hơn Stemming nhưng kết quả luôn là một từ có nghĩa và chính xác.
- *Lưu ý với Tiếng Việt:* Tiếng Việt là ngôn ngữ đơn lập, các từ không thay đổi hình thái theo thì/ngôi/số. Do đó, cả Stemming và Lemmatization thường không mang lại ý nghĩa và ít được áp dụng trong NLP tiếng Việt. Thay vào đó, ta tập trung vào quá trình Tách từ ghép (Word Segmentation).

### Giải thích TF-IDF và sự khác biệt

**1. CountVectorizer — Bag of Words (BoW)**
- **Công thức:** Đếm tần số từ: `count(word, document)`
- **Ví dụ:** "tích cực tích cực tốt" → `{'tích cực': 2, 'tốt': 1}`
- **Nhược điểm:**
  - Từ thường (stopwords) có tần số cao nhưng ít ý nghĩa ("và", "là", "cái")
  - Không phân biệt từ quan trọng vs từ bình thường
  - Mô hình có thể bị "lôi kéo" bởi những từ phổ biến nhưng không mang ý nghĩa

**2. TfidfVectorizer — TF-IDF Weighting**
- **TF (Term Frequency):** Tần số từ trong document: $\text{TF}(t, d) = \frac{\text{count}(t, d)}{\text{total words in } d}$
- **IDF (Inverse Document Frequency):** Độ hiếm của từ trong toàn bộ corpus: $\text{IDF}(t) = \log\left(\frac{\text{total documents}}{\text{documents containing } t}\right)$
- **TF-IDF:** $\text{TF-IDF}(t, d) = \text{TF}(t, d) \times \text{IDF}(t)$

**3. Ưu điểm của TF-IDF**
- Từ phổ biến (stopwords) → IDF thấp → TF-IDF thấp → ít ảnh hưởng
- Từ hiếm nhưng xuất hiện nhiều trong 1 document → TF-IDF cao → phân biệt tốt hơn
- Giảm overfitting do loại bỏ "nhiễu" từ thường
- n-gram (ngram_range=(1,2)): Bắt được đặc điểm cụm từ (bi-grams) như "rất tệ", "rất tốt"

**4. Khi nào dùng gì?**

| Vectorizer | Khi nào dùng | Ưu điểm | Nhược điểm |
|------------|-------------|--------|-----------|
| **CountVectorizer** | Dữ liệu nhỏ, model đơn giản, Naive Bayes | Nhanh, đơn giản | Bị ảnh hưởng từ thường, tần số không chuẩn hóa |
| **TfidfVectorizer** | Dữ liệu lớn, cần hiệu năng cao, SVM/LR | Chuẩn hóa, loại bỏ từ thường, có n-gram | Tính toán phức tạp hơn, khó đọc kết quả |

**Kết luận cho bài này:** Dùng **TfidfVectorizer(max_features=10000, ngram_range=(1,2))** vì:
- Dataset có từ thường tiếng Việt nhiều (và, là, cái,...)
- N-gram bắt được cụm từ có ý nghĩa ("rất tích cực", "hàng kém")
- Phù hợp với các model SVM, Logistic Regression yêu cầu đặc trưng chuẩn hóa

### A6 — Data Leakage: Fitting vectorizer sai cách

**Data Leakage là gì?**  
Khi thông tin từ test set "rò rỉ" vào quá trình huấn luyện model, làm cho kết quả evaluation **lạc quan không thực tế**.

**Sai lầm phổ biến: Fit vectorizer trên test set**

```python
# SAI LẦM (DATA LEAKAGE)
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.2, stratify=df['label'], random_state=42
)

# FIT trên toàn bộ data (bao gồm test) → IDF được tính từ test!
vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_vec = vectorizer.fit_transform(X_train + X_test)  # LỖI!
X_test_vec = vectorizer.transform(X_test)

# Hậu quả: IDF của từ trong test được "nhìn" trước → model học "gian lận"
```

**Vấn đề cụ thể:**
1. **IDF bị "sâu":** Nếu từ "quả_bom" chỉ xuất hiện trong test, nhưng khi fit trên train+test, IDF của nó được tính sẽ khác so với khi fit riêng trên train.
2. **Model lạc quan:** Khi evaluation trên test, model đạt accuracy cao hơn thực tế vì nó đã "biết" về từ vựng test từ trước.
3. **Validation set bị "ô nhiễm":** Nếu vectorizer fit trên train+val+test, cả val cũng bị lạc quan.

**CÁCH ĐÚNG:**

```python
# ĐÚNG (Không có data leakage)
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.2, stratify=df['label'], random_state=42
)

# FIT chỉ trên train set
vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_vec = vectorizer.fit_transform(X_train)  # Fit riêng trên train

# TRANSFORM test set (không fit lại!)
X_test_vec = vectorizer.transform(X_test)  # Chỉ transform, không fit

# Kết quả evaluation sẽ thực tế hơn
```

**Tóm lại (quy tắc vàng):**
- FIT (fit_transform): Chỉ dùng **trên train set duy nhất**
- TRANSFORM: Áp dụng lên val/test set
- **Công thức:** `vectorizer.fit_transform(X_train)` → `vectorizer.transform(X_val)` → `vectorizer.transform(X_test)`

````python
from sklearn.model_selection import train_test_split

print("\n" + "=" * 70)
print("A6 — VÍ DỤ: CÁCH ĐÚNG để dùng Vectorizer (KHÔNG DATA LEAKAGE)")
print("=" * 70)

# 1. Split data an toàn
X = df['text']
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print(f"\n[1] Split dữ liệu:")
print(f"  Train: {len(X_train)} mẫu")
print(f"  Test: {len(X_test)} mẫu")

# 2. Fit vectorizer CHỈ trên train
print(f"\n[2] Fit TfidfVectorizer CHỈ trên train set:")
tfidf_correct = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_tfidf = tfidf_correct.fit_transform(X_train)
print(f"  Fit done. Shape: {X_train_tfidf.shape}")

# 3. Transform test (KHÔNG fit lại)
print(f"\n[3] Transform test set (không fit lại):")
X_test_tfidf = tfidf_correct.transform(X_test)
print(f"  Transform done. Shape: {X_test_tfidf.shape}")

# 4. So sánh: Nếu fit sai cách (data leakage)
print(f"\n[4] So sánh với cách sai (fit trên train + test):")
tfidf_wrong = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_wrong = tfidf_wrong.fit_transform(
    pd.concat([X_train, X_test])  # Fit trên cả train + test
)
X_test_wrong = tfidf_wrong.transform(X_test)

# Lấy IDF của một vài từ để so sánh
sample_features = ['tốt', 'xấu', 'tích_cuc']
print(f"\n  So sánh IDF của một vài từ:")
print(f"  {'Từ':<15} {'IDF (Đúng)':<15} {'IDF (Sai)':<15}")
print(f"  {'-'*45}")

feature_names_correct = tfidf_correct.get_feature_names_out()
feature_names_wrong = tfidf_wrong.get_feature_names_out()

for word in sample_features:
    # Tìm index của từ trong từng vectorizer
    if word in feature_names_correct:
        idf_correct = tfidf_correct.idf_[list(feature_names_correct).index(word)]
    else:
        idf_correct = "Không có"

    if word in feature_names_wrong:
        idf_wrong = tfidf_wrong.idf_[list(feature_names_wrong).index(word)]
    else:
        idf_wrong = "Không có"

    print(f"  {word:<15} {str(idf_correct):<15} {str(idf_wrong):<15}")

print(f"\nLưu ý: IDF khác nhau → từ sẽ được trọng số hóa khác nhau!")
print(f"   Model training sẽ học các trọng số khác nhau → dự đoán khác nhau!")
print(f"\nKẾT LUẬN: Luôn fit vectorizer CHỈ trên train set, sau đó transform val/test")
````

## C2 — MultinomialNB & LogisticRegression (grid hyperparameter)

**Yêu cầu đề bài:**
- **MultinomialNB:** `alpha` ∈ {0.1, 0.5, 1.0}
- **LogisticRegression:** `C` ∈ {0.1, 1, 10}, `solver='lbfgs'`, `max_iter=1000`
- Features: TF-IDF (`max_features=10000`, `ngram_range=(1,2)`) — fit **chỉ trên train**
- Chọn hyperparameter tốt nhất dựa trên **validation accuracy**

````python
# [Bình] C2 — Chuẩn bị dữ liệu + grid search trên validation set

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

RANDOM_STATE = 42

X = df["text_clean"]
y = df["label"]

# Split 70/10/20 (stratify, random_state cố định)
X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval, test_size=0.125, random_state=RANDOM_STATE, stratify=y_trainval
)

print("Phân phối nhãn từng tập:")
for name, labels in [("Train", y_train), ("Val", y_val), ("Test", y_test)]:
    print(f"\n{name} ({len(labels):,} mẫu):")
    print(labels.value_counts(normalize=True).mul(100).round(1).astype(str) + "%")

# TF-IDF: fit chỉ trên train, transform val/test
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_vec = tfidf.fit_transform(X_train)
X_val_vec = tfidf.transform(X_val)
X_test_vec = tfidf.transform(X_test)

print(f"\nTF-IDF shape — train: {X_train_vec.shape}, val: {X_val_vec.shape}, test: {X_test_vec.shape}")


def grid_search_on_val(model_class, param_name, param_values, fixed_params=None):
    fixed_params = fixed_params or {}
    rows = []
    best_model, best_acc, best_value = None, -1.0, None

    for value in param_values:
        params = {param_name: value, **fixed_params}
        model = model_class(**params)
        model.fit(X_train_vec, y_train)
        val_acc = accuracy_score(y_val, model.predict(X_val_vec))
        rows.append({param_name: value, "val_accuracy": val_acc})
        if val_acc > best_acc:
            best_acc, best_value, best_model = val_acc, value, model

    return pd.DataFrame(rows), best_model, best_value, best_acc


# --- MultinomialNB ---
nb_grid, best_nb, best_nb_alpha, best_nb_acc = grid_search_on_val(
    MultinomialNB, "alpha", [0.1, 0.5, 1.0]
)
print("=== MultinomialNB — grid alpha ===")
display(nb_grid)
print(f"Best: alpha={best_nb_alpha}, val accuracy={best_nb_acc:.4f}")

# --- LogisticRegression ---
lr_grid, best_lr, best_lr_C, best_lr_acc = grid_search_on_val(
    LogisticRegression,
    "C",
    [0.1, 1, 10],
    fixed_params={"solver": "lbfgs", "max_iter": 1000, "random_state": RANDOM_STATE},
)
print("\n=== LogisticRegression — grid C ===")
display(lr_grid)
print(f"Best: C={best_lr_C}, val accuracy={best_lr_acc:.4f}")
````

### Trả lời câu hỏi C2 — Tham số `C` của Logistic Regression & overfitting

**`C` là gì?**  
`C` là **nghịch đảo độ mạnh regularization** (L2). Càng **lớn** → penalty càng **nhỏ** → model **linh hoạt hơn**; càng **nhỏ** → regularization càng **mạnh** → model **đơn giản hơn**.

| Giá trị `C` | Regularization | Hành vi | Rủi ro |
|---|---|---|---|
| **0.1** (nhỏ) | Mạnh | Trọng số bị ép nhỏ, biên quyết định mượt | Có thể **underfit** nếu quá nhỏ |
| **1** | Trung bình | Cân bằng bias–variance | Thường là điểm khởi đầu tốt |
| **10** (lớn) | Yếu | Model bám sát train hơn | Dễ **overfit** trên TF-IDF nhiều chiều |

**Liên hệ overfitting:** Với bài này, TF-IDF tạo tới 10.000 features trong khi class tiêu cực chỉ ~354 mẫu. Nếu chọn `C` quá lớn, Logistic Regression có thể ghi nhớ nhiễu trên train/val (accuracy val cao nhưng generalize kém trên test). Chọn `C` nhỏ hơn giúp **giảm overfitting** bằng cách phạt trọng số lớn — đó là lý do ta grid search trên validation thay vì mặc định `C=10`.

## C3 — Stratified K-Fold Cross-Validation (k=5) cho NB & LR

**Mục tiêu:** Đánh giá độ ổn định của 2 mô hình bằng **StratifiedKFold (k=5)** trên tập train+val (90% dữ liệu), báo cáo **mean ± std accuracy**.

Dùng `Pipeline` (TF-IDF + classifier) để **không data leakage** — vectorizer được fit lại trong từng fold.

````python
# [Bình] C3 — Stratified K-Fold CV (k=5)

from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

models = {
    "MultinomialNB": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=10000, ngram_range=(1, 2))),
        ("clf", MultinomialNB(alpha=best_nb_alpha)),
    ]),
    "LogisticRegression": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=10000, ngram_range=(1, 2))),
        ("clf", LogisticRegression(
            C=best_lr_C, solver="lbfgs", max_iter=1000, random_state=RANDOM_STATE
        )),
    ]),
}

cv_rows = []
print("=== Stratified 5-Fold CV (train+val, accuracy) ===\n")
for name, pipeline in models.items():
    scores = cross_val_score(
        pipeline, X_trainval, y_trainval, cv=skf, scoring="accuracy", n_jobs=-1
    )
    cv_rows.append({
        "Model": name,
        "Mean Accuracy": scores.mean(),
        "Std": scores.std(),
    })
    print(f"{name}: {scores.mean():.4f} ± {scores.std():.4f}")
    print(f"  Fold scores: {[round(s, 4) for s in scores]}\n")

cv_df = pd.DataFrame(cv_rows)
display(cv_df)
````

# CÔNG VIỆC NGÀY 4 CỦA HẢI

### C2 — Huấn luyện LinearSVC & RandomForest (Hải)

**Mục tiêu:** Thực hiện tìm kiếm tham số tối ưu (Grid Search) cho 2 mô hình còn lại trên tập Validation.

````python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline

# Khởi tạo lại các biến dữ liệu từ phần của Bình để đảm bảo tồn tại trong kernel
_RS = 42
X = df['text_clean'] if 'text_clean' in df.columns else df['text']
y = df['label']

X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, test_size=0.2, random_state=_RS, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.125, random_state=_RS, stratify=y_trainval)

tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_vec = tfidf.fit_transform(X_train)
X_val_vec = tfidf.transform(X_val)
X_test_vec = tfidf.transform(X_test)

def grid_search_on_val_fixed(model_class, param_name, param_values, xtrain, ytrain, xval, yval, fixed_params=None):
    fixed_params = fixed_params or {}
    rows = []
    best_model, best_acc, best_value = None, -1.0, None
    for value in param_values:
        params = {param_name: value, **fixed_params}
        model = model_class(**params)
        model.fit(xtrain, ytrain)
        val_acc = accuracy_score(yval, model.predict(xval))
        rows.append({param_name: value, "val_accuracy": val_acc})
        if val_acc > best_acc:
            best_acc, best_value, best_model = val_acc, value, model
    return pd.DataFrame(rows), best_model, best_value, best_acc

# --- 1. LinearSVC Grid Search ---
svc_grid, best_svc, best_svc_C, best_svc_acc = grid_search_on_val_fixed(
    LinearSVC, "C", [0.1, 1, 10], X_train_vec, y_train, X_val_vec, y_val,
    fixed_params={"max_iter": 2000, "random_state": _RS}
)
print("=== LinearSVC — grid C ===")
display(svc_grid)

# --- 2. RandomForest Grid Search (n_estimators × max_depth) ---
from sklearn.model_selection import ParameterGrid

rf_rows = []
best_rf, best_rf_acc = None, -1.0
best_rf_n_estimators, best_rf_depth = 100, None
for params in ParameterGrid({"n_estimators": [100, 200], "max_depth": [None, 10, 20]}):
    model = RandomForestClassifier(
        n_estimators=params["n_estimators"],
        max_depth=params["max_depth"],
        random_state=_RS,
        n_jobs=-1,
    )
    model.fit(X_train_vec, y_train)
    val_acc = accuracy_score(y_val, model.predict(X_val_vec))
    rf_rows.append({
        "n_estimators": params["n_estimators"],
        "max_depth": params["max_depth"],
        "val_accuracy": val_acc,
    })
    if val_acc > best_rf_acc:
        best_rf_acc = val_acc
        best_rf = model
        best_rf_n_estimators = params["n_estimators"]
        best_rf_depth = params["max_depth"]

rf_grid = pd.DataFrame(rf_rows)
print("\n=== RandomForest — grid n_estimators × max_depth ===")
display(rf_grid.sort_values("val_accuracy", ascending=False))
print(f"Best RF: n_estimators={best_rf_n_estimators}, max_depth={best_rf_depth}, val acc={best_rf_acc:.4f}")
````

### Trả lời lý thuyết C2: Margin/Support Vectors & Overfitting (Hải)

**1. Margin và Support Vectors (SVM):**
- **Margin:** Là khoảng cách giữa đường biên quyết định (decision boundary) và các điểm dữ liệu gần nó nhất. SVM cố gắng tối đa hóa khoảng cách này để tăng khả năng tổng quát hóa.
- **Support Vectors:** Là những điểm dữ liệu nằm ngay sát biên hoặc vi phạm biên. Chúng là những nhân tố quan trọng nhất xác định vị trí và hướng của đường biên.

**2. Tại sao Random Forest (RF) chống Overfitting tốt?**
- RF sử dụng kỹ thuật **Bagging** (Bootstrap Aggregating) và chọn ngẫu nhiên tập con các đặc trưng (features) cho mỗi nút chia.
- Việc kết hợp (ensemble) kết quả từ nhiều cây quyết định độc lập giúp triệt tiêu các sai số ngẫu nhiên của từng cây đơn lẻ, làm giảm phương sai (variance) của toàn hệ thống.

### C3 — Stratified 5-Fold Cross-Validation cho SVM & RF (Hải)

````python
# C3 — Stratified 5-Fold Cross-Validation cho SVM & RF (Hải)
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Tái sử dụng tham số đã tìm được
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

hai_models = {
    "LinearSVC": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=10000, ngram_range=(1, 2))),
        ("clf", LinearSVC(C=best_svc_C, max_iter=2000, random_state=42)),
    ]),
    "RandomForest": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=10000, ngram_range=(1, 2))),
        ("clf", RandomForestClassifier(n_estimators=best_rf_n_estimators, max_depth=best_rf_depth, random_state=42, n_jobs=-1)),
    ]),
}

hai_cv_rows = []
print("=== Stratified 5-Fold CV (Hải's Models) ===\n")
for name, pipeline in hai_models.items():
    scores = cross_val_score(pipeline, X_trainval, y_trainval, cv=skf, scoring='accuracy', n_jobs=-1)
    hai_cv_rows.append({
        "Model": name,
        "Mean Accuracy": scores.mean(),
        "Std": scores.std(),
    })
    print(f"{name}: {scores.mean():.4f} ± {scores.std():.4f}")

hai_cv_df = pd.DataFrame(hai_cv_rows)
display(hai_cv_df)
````

## D1 — Bảng metrics trên test set (4 models) [Hải]

Đánh giá **Accuracy, Precision, Recall, F1 (macro)** cho 4 model đã chọn hyperparameter tốt nhất trên validation (C2).

````python
# [Hải] D1 — Metrics test cho 4 models

from sklearn.metrics import precision_recall_fscore_support, classification_report

# Đảm bảo test đã vectorize (Hai C2 có thể chưa set X_test_vec)
if "X_test_vec" not in globals():
    X_test_vec = tfidf.transform(X_test)

models_test = {
    "MultinomialNB": best_nb,
    "LogisticRegression": best_lr,
    "LinearSVC": best_svc,
    "RandomForest": best_rf,
}

d1_rows = []
for name, model in models_test.items():
    y_pred = model.predict(X_test_vec)
    acc = accuracy_score(y_test, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(
        y_test, y_pred, average="macro", zero_division=0
    )
    d1_rows.append({
        "Model": name,
        "Accuracy": round(acc, 4),
        "Precision (macro)": round(prec, 4),
        "Recall (macro)": round(rec, 4),
        "F1 (macro)": round(f1, 4),
    })

metrics_test_df = pd.DataFrame(d1_rows).sort_values("F1 (macro)", ascending=False)
display(metrics_test_df)

best_model_name = metrics_test_df.iloc[0]["Model"]
best_model = models_test[best_model_name]
print(f"\n→ Model tốt nhất trên test (F1 macro): {best_model_name}")
````

- **`tieu_cuc`:** **Recall thấp** là rủi ro lớn nhất — bỏ sót review tiêu cực dù Accuracy tổng vẫn ~85%.

**Tình huống lọc review tiêu cực (moderation):**

## E1 — GridSearchCV (≥2 model, ≥3 tham số) [Hải]

GridSearchCV trên **train+val** (90% data), `StratifiedKFold(k=5)`, metric `f1_macro` — tìm hyperparameter tốt hơn grid thủ công C2.

````python
# [Hải] E1 — GridSearchCV cho MultinomialNB & LogisticRegression

from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline

cv_tune = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

pipelines = {
    "MultinomialNB": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=10000, ngram_range=(1, 2))),
        ("clf", MultinomialNB()),
    ]),
    "LogisticRegression": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=10000, ngram_range=(1, 2))),
        ("clf", LogisticRegression(solver="lbfgs", max_iter=1000, random_state=RANDOM_STATE)),
    ]),
}

param_grids = {
    "MultinomialNB": {
        "clf__alpha": [0.05, 0.1, 0.5, 1.0, 2.0],
    },
    "LogisticRegression": {
        "clf__C": [0.01, 0.1, 1, 10, 100],
    },
}

e1_results = []
for name, pipe in pipelines.items():
    gs = GridSearchCV(
        pipe,
        param_grids[name],
        cv=cv_tune,
        scoring="f1_macro",
        n_jobs=-1,
        refit=True,
    )
    gs.fit(X_trainval, y_trainval)
    e1_results.append({
        "Model": name,
        "Best params": gs.best_params_,
        "Best CV F1 macro": round(gs.best_score_, 4),
        "Estimator": gs.best_estimator_,
    })
    print(f"{name}: best={gs.best_params_}, CV F1 macro={gs.best_score_:.4f}")

e1_df = pd.DataFrame([{k: v for k, v in r.items() if k != "Estimator"} for r in e1_results])
display(e1_df)

# So sánh trước/sau (val accuracy thủ công C2 vs CV F1 sau GridSearch)
print("\n=== So sánh tuning C2 (val acc) vs E1 (CV F1 macro) ===")
before = {
    "MultinomialNB": best_nb_acc,
    "LogisticRegression": best_lr_acc,
}
for r in e1_results:
    print(f"{r['Model']}: val acc C2={before[r['Model']]:.4f} → GridSearch CV F1={r['Best CV F1 macro']:.4f}")
````

### E2 — Câu hỏi tuning [Hải]

**GridSearch vs RandomizedSearch:** GridSearch thử **hết** tổ hợp trong lưới → chính xác nhưng **chậm** khi lưới lớn. RandomizedSearch thử **ngẫu nhiên** N tổ hợp → nhanh hơn, phù hợp RF/SVM nhiều tham số.

**Tại sao CV trong tuning?** Một lần chia val có thể **may/rủi**; CV (5-fold) lấy **trung bình nhiều lần chia** → ước lượng ổn định hơn, giảm overfit vào một split val cố định.

**Diminishing returns:** Sau khi params “đủ tốt”, thêm giá trị grid / fold chỉ cải thiện **vài phần nghìn** F1 nhưng tốn CPU gấp bội — nên dừng khi metric plateau (như CV F1 ~0.55–0.65 macro trên data lệch class).

## E3 — Cải tiến: `class_weight='balanced'` [Hải]

So sánh Logistic Regression **trước / sau** khi cân bằng trọng số lớp thiểu số (`tieu_cuc`, `trung_lap`).

````python
# [Hải] E3 — class_weight='balanced' vs mặc định

from sklearn.metrics import f1_score

def eval_lr(class_weight):
    model = LogisticRegression(
        C=best_lr_C,
        solver="lbfgs",
        max_iter=1000,
        random_state=RANDOM_STATE,
        class_weight=class_weight,
    )
    model.fit(X_train_vec, y_train)
    pred = model.predict(X_test_vec)
    return {
        "class_weight": "balanced" if class_weight else "None (default)",
        "Accuracy": round(accuracy_score(y_test, pred), 4),
        "F1 macro": round(f1_score(y_test, pred, average="macro"), 4),
        "Recall tieu_cuc": round(
            precision_recall_fscore_support(
                y_test, pred, labels=["tieu_cuc"], average=None, zero_division=0
            )[1][0],
            4,
        ),
    }

e3_df = pd.DataFrame([
    eval_lr(None),
    eval_lr("balanced"),
])
display(e3_df)

print("\nNhận xét: class_weight='balanced' thường tăng Recall tieu_cuc / F1 macro")
print("đổi lại Accuracy tổng có thể giảm nhẹ — phù hợp bài toán ưu tiên bắt review tiêu cực.")
````

## D4 — Phản biện: "Accuracy ~85% đã đủ để đưa vào production?"

**Câu hỏi:** Với accuracy validation/test khoảng **85–86%** (MultinomialNB / LogisticRegression trên Tiki reviews), liệu mô hình đã sẵn sàng triển khai thực tế?

**Luận điểm: Chưa đủ — accuracy cao có thể gây hiểu nhầm.**

Thứ nhất, dataset **mất cân bằng nặng** (~85,6% `tich_cuc`, ~2,6% `tieu_cuc`). Một model “lười” luôn đoán tích cực vẫn có thể đạt accuracy ~86% mà **không bắt được review tiêu cực** — đúng use-case quan trọng nhất (lọc khiếu nại, cảnh báo seller). Accuracy là **trung bình có trọng số theo tần suất lớp**, không phản ánh chất lượng trên lớp thiểu số.

Thứ hai, bài toán sản phẩm thường **bất đối xứng chi phí lỗi**. Bỏ sót review tiêu cực (false negative) có thể để lọt nội dung xấu; gắn nhầm tiêu cực (false positive) có thể ẩn review trung thực. Cần báo cáo **Precision / Recall theo từng nhãn** (Phần D2) và chọn ngưỡng theo mục tiêu kinh doanh, không dừng ở một con số accuracy.

Thứ tư, hạn chế mô hình: TF-IDF + linear/NB **không nắm ngữ cảnh** (mỉa, phủ định, từ lóng). Misclassified (D3) cho thấy nhầm **trung_lap ↔ tich_cuc** và **tieu_cuc → tich_cuc** — đúng pattern khi thiếu ngữ nghĩa sâu.

**Kết luận:** 85% accuracy là **baseline học thuật chấp nhận được**, nhưng **chưa đủ** để coi là production-ready cho moderation sentiment. Nên yêu cầu thêm: **F1 macro**, recall `tieu_cuc`, confusion matrix, calibration, và quy trình review thủ công cho case confidence thấp. Chỉ khi metric lớp thiểu số đạt ngưỡng nghiệp vụ mới nên triển khai rộng.
